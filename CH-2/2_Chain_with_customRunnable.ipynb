```python
from langchain_core.runnables import RunnableLambda
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

# Model
llm = ChatOllama(model="phi3:latest")

# Prompt
prompt = PromptTemplate.from_template(
    "Explain {topic} in simple words."
)

# Custom runnable function
def make_uppercase(inputs):
    inputs["topic"] = inputs["topic"].upper()
    return inputs

uppercase_step = RunnableLambda(make_uppercase)

# Build chain
chain = uppercase_step | prompt | llm

# Run
response = chain.invoke({"topic": "docker"})

print(response.content)
```
