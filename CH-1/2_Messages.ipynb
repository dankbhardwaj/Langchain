```python
from langchain_ollama import ChatOllama
from langchain_core.messages import SystemMessage, HumanMessage

llm = ChatOllama(model="llama3:latest")

messages = [
    SystemMessage(content="You are a senior DevOps engineer. Give structured answers."),
    HumanMessage(content="Explain Kubernetes in simple words.")
]

response = llm.invoke(messages)

print(response.content)
messages.append(HumanMessage(content="How is it different from Docker?"))

response = llm.invoke(messages)

print(response.content)
```

```python
from langchain_ollama import OllamaLLM

llm = OllamaLLM(
    model="phi3:latest",
    num_predict=60
)

print(llm.invoke("Say hello in one short sentence."))
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_ollama import OllamaLLM

# Create model
llm = OllamaLLM(
    model="phi3:latest",
    num_predict=100
)

# Create template
template = """
You are a senior DevOps engineer.

Explain the following topic clearly:
{topic}

Give:
1. Simple definition
2. Real-world example
3. Why it is important
"""

prompt = PromptTemplate.from_template(template)

# Create chain
chain = prompt | llm

# Run
response = chain.invoke({"topic": "Kubernetes"})

print(response)
print(chain.invoke({"topic": "Docker"}))
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="phi3:latest",
    num_predict=100
)

chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a senior DevOps engineer."),
    ("human", "Explain {topic} in simple terms.")
])

chain = chat_prompt | llm

response = chain.invoke({"topic": "Ci cd pipelines"})

print(response.content)
```
